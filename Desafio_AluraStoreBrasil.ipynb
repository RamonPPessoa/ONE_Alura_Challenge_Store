### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
faturametento_total = loja['Preço'].sum()
faturametento_total
f"R$ {loja['Preço'].sum():.2f}"
```

```python
faturametento_total2 = loja2['Preço'].sum()
faturametento_total2
f"R$ {loja2['Preço'].sum():.2f}"
```

```python
faturametento_total3 = loja3['Preço'].sum()
faturametento_total3
f"R$ {loja3['Preço'].sum():.2f}"
```

```python
faturametento_total4 = loja4['Preço'].sum()
faturametento_total4
f"R$ {loja4['Preço'].sum():.2f}"
```

```python
faturamento = {
    'loja 1': faturametento_total,
    'loja 2': faturametento_total2,
    'loja 3': faturametento_total3,
    'loja 4': faturametento_total4
}

lojas_faturamento = pd.DataFrame(list(faturamento.items()), columns=['Loja', 'Faturamento'])

# Gráfico de colunas horizontais
lojas_faturamento.plot(
    x='Loja', y='Faturamento',
    xlabel='Faturamento', ylabel='Lojas',
    kind='barh', figsize=(10,5), color='blue'
);

# Adicionando os valores em cada barra
ax = plt.gca()  # pega o eixo atual do gráfico
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', label_type='edge')

plt.show()
```

# 2. Vendas por Categoria

```python
vendas_por_categoria2 = loja2['Categoria do Produto'].value_counts()
vendas_por_categoria2
```

```python
vendas_por_categoria3 = loja3['Categoria do Produto'].value_counts()
vendas_por_categoria3
```

```python
vendas_por_categoria4 = loja4['Categoria do Produto'].value_counts()
vendas_por_categoria4
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suas variáveis já calculadas
categoria1 = vendas_por_categoria
categoria2 = vendas_por_categoria2
categoria3 = vendas_por_categoria3
categoria4 = vendas_por_categoria4

# Criando DataFrame com todas as lojas
df_vendas = pd.DataFrame({
    'Loja 1': categoria1,
    'Loja 2': categoria2,
    'Loja 3': categoria3,
    'Loja 4': categoria4
}).fillna(0)

# Transpondo para ter lojas como linhas e categorias como colunas
df_vendas_transposto = df_vendas.T

# Gráfico de barras empilhadas
ax = df_vendas_transposto.plot(
    kind='bar',
    stacked=True,
    figsize=(14, 7),
    colormap='tab20'
)

plt.title('Totais de Vendas por Categoria dentro de cada Loja')
plt.xlabel('Lojas')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=0)
plt.legend(title='Categorias de Produtos', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adicionando os valores em cada segmento da barra
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', label_type='center')

plt.tight_layout()
plt.show()

```

##Introdução

Este relatório apresenta a análise das quatro lojas fictícias da rede Alura Store, com base nas bases de dados em formato CSV disponibilizadas pelo Senhor João, que contêm informações de faturamento, categorias de produtos, avaliações dos clientes, desempenho de vendas e frete médio. O objetivo é identificar a loja com menor eficiência e recomendar qual deve ser vendida para que o Senhor João possa iniciar um novo empreendimento.

Foram avaliados os seguintes critério:

1.   Faturamento
2.   Vendas por categoria
3.   Média de avaliação dos clientes
4.   Produtos mais e menos vendidos



## 📈 Análises
**1.** Faturamento

  * Loja 1: R$ 1.534.090,12

  * Loja 2: R$ 1.489.854,06

  * Loja 3: R$ 1.464.025,03

  * Loja 4: R$ 1.384.497,58

➡️ Loja 1 tem o maior faturamento, enquanto Loja 4 apresenta o menor.

**2.** Vendas por Categoria


*   Loja 3 e Loja 4 se destacam em Utilidades Domésticas e categorias estratégicas.
*   Loja 1 apresenta mix equilibrado, mas sem liderança clara.

**3.** Média de Avaliação dos Clientes


*   Loja 1: 3.98 (pior avaliação)
*   Loja 2: 4.04
*  Loja 3: 4.05 (melhor avaliação)
*   Loja 4: 4.00

➡️ Loja 1 é a menos eficiente em satisfação dos clientes.




**4.** Produtos Mais e Menos Vendidos



*   Mais vendidos: Loja 3 e Loja 4 aparecem com força nos campeões de venda (ex.: Bicicleta, TV Led UHD 4K, Kit ferramentas).
*   Menos vendidos: Loja 1 aparece com frequência entre os encalhados, mostrando mix pouco eficiente.

**5.** Frete Médio


*   Loja 1: R$ 34,69 (mais caro)

*   Loja 2: R$ 33,62

*   Loja 3: R$ 33,07

*   Loja 4: R$ 31,28 (mais barato)

➡️ Loja 1 tem o frete mais caro, impactando negativamente na margem de lucro.

##🔎 Interpretação Geral


*   **Loja 1:** Apesar do maior faturamento, **apresenta pior avaliação, frete mais caro, menos destaque nos campeões de venda e presença constante entre os encalhados**. É a **loja menos eficiente** no conjunto dos critérios.
*   **Loja 4:** Tem o menor faturamento, **mas compensa com frete mais barato, boa presença nos produtos mais vendidos e força em categorias estratégicas.**
*  **Loja 3:** Melhor avaliação e bom desempenho nos mais vendidos, mas faturamento intermediário.
*  **Loja 2:** Desempenho equilibrado, sem grandes pontos negativos.
##✅ Recomendação
Com base nos cinco critérios, a loja que apresenta** maior ineficiência é a Loja 1**. Portanto, recomenda-se que o Senhor João **venda a Loja 1** para iniciar seu novo empreendimento.

##🚀 Potencial de Fortalecimento da Loja 4
Embora a Loja 4 tenha o menor faturamento, ela possui vantagens que podem ser exploradas para aumentar sua eficiência:


*   **Frete competitivo:** já é o mais barato, pode ser usado como diferencial de marketing.
*   **Produtos campeões de venda:** reforçar campanhas em torno de Bicicleta, TV Led UHD 4K e Kit ferramentas.

*   **Categorias fortes: **expandir ainda mais em Móveis e Utilidades Domésticas, onde já se destaca.
*   **Melhoria na avaliação:** investir em atendimento ao cliente, programas de fidelização e promoções para elevar a nota de satisfação.
*  **Gestão de estoque:** reduzir produtos encalhados com promoções e kits combinados.

Com essas estratégias, a Loja 4 pode compensar seu faturamento menor e se tornar uma operação mais eficiente e lucrativa.

##Conclusão
A análise completa mostra que a **Loja 1 é a menos eficiente** e deve ser vendida. Ao mesmo tempo, **a Loja 4 tem potencial de crescimento** e pode ser fortalecida com estratégias de marketing, fidelização e otimização de estoque, garantindo vantagens competitivas para o futuro da rede.
